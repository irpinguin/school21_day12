2. Размер figsize остается прежним, размер шрифта можно выбрать по своему вкусу.
3. Цветовая палитра должна быть такой же, как в примере.
4. Размер шрифта заголовка — 15.
5. Ширина линий ящиков — 3, ширина линий медианы — 2.


```python
df.boxplot(column=['after', 'before'],
        by='group', 
        fontsize=10, 
        figsize=(15, 8),
        grid=True,
        showmeans=True,
        boxprops = dict(linestyle='-', linewidth=3, color='black'),
        medianprops = dict(linestyle='-', linewidth=2, color='red'),
        whiskerprops = dict(linestyle='-', linewidth=1, color='red'),
        capprops = dict(linestyle='-', linewidth=1, color='red'),
          )
plt.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
```
