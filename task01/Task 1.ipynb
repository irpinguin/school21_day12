```python
import pandas as pd
from sqlite3 import connect
```

**1. Установите соединение с базой данных (как и в прошлом проекте).** 

```python
con = connect('/content/drive/MyDrive/School21/day12/data/checking-logs.sqlite')
```

**2. Выполните запрос, который достает datetime из таблицы pageviews; выбирайте только пользователей, а не администраторов.**  

```python
views = pd.io.sql.read_sql('SELECT datetime\
                            FROM pageviews \
                            WHERE uid LIKE "user__%" ',
                            con, parse_dates = ['datetime'])
views
```

**3. Используя библиотеку Pandas, создайте новый датафрейм, в котором посещения страницы будут подсчитаны и сгруппированы по дате.**  

```python
views['date'] = views['datetime'].dt.date
views['views'] = 1
```

```python
views_number = views.drop(columns="datetime").groupby(by='date', dropna = False).count()
views_number.head()
```

```python
views_number.shape
```


**4. Используя метод .plot() библиотеки Pandas, создайте график.**  
*   **Размер шрифта — 8.**
*   **figsize — 15,8.**
*   **Название графика — «Число просмотров в день».**
*   **Обратите внимание на поворот надписей на оси x на графике ниже.**  

```python
views_number.plot(figsize=(15, 8), rot=90, fontsize=8,
                  title="Views per day",
                  xlabel='Date') 
pass
```
